```python
#making it more standard



def solve(director_name, start_year, end_year):

  df = directors.loc[directors["director_name"] == director_name]

  if len(df) != 0:

    director_id = df.reset_index(drop=True)['id'][0]

    director_movie= movies.loc[movies['director_id']== director_id]

    res= director_movie[(director_movie['year']>=start_year) & (director_movie['year']<=end_year)]

    return res

  else:

    return "Director Not Found!"
```

# **Q9 - In which decade, the most number of movies were produced?**

```python
data.describe()
```

```python
year_bin = [1970, 1980, 1990, 2000, 2010, 2020]
bin_labels = ["A", "B", "C", "D", "E"]
```

```python
data["year_bucket"] = pd.cut(data["year"], bins = year_bin, labels = bin_labels)
data.head()
```

```python
data["year_bucket"].value_counts()
```

I performed analytics on the movie dataset by creating year-based bins to represent different decades and assigned appropriate labels.
I then calculated the frequency distribution using the value_counts() function to identify which decade had the highest number of movies produced.

Our analysis indicates that the **2000–2010 decade** experienced the highest volume of movie production. This suggests a significant expansion phase in the film industry during that period.
